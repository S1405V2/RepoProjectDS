```python
def data_array(url):
    request_text = request.urlopen(url).read()
    page = bs4.BeautifulSoup(request_text, "lxml")
    array = page.find('table', {'class': 'c-table'})
    table_body = array.find('tbody')
    rows = table_body.findAll('tr')
    cols = [rows[i].find_all('td') for i in range(len(rows))]
    final_data = np.array([[x[i].text.strip() for i in range(len(x))] for x in cols])
    data = pd.DataFrame(final_data, columns = l)
    return data
```
