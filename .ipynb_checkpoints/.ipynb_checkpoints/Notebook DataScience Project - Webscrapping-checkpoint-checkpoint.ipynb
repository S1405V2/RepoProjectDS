```python
import urllib
import bs4
import pandas as pd
from urllib import request
import numpy as np
import csv
```

```python
url_btc_usd = "https://www.boursorama.com/bourse/devises/cours/historique/9xXBTUSDSPOT?fromSymbol=BTC&toSymbol=USD&fromLabel=bitcoin&toLabel=dollar"
url_eth_usd = "https://www.boursorama.com/bourse/devises/cours/historique/9xETHUSDSPOT?fromSymbol=ETH&toSymbol=USD&fromLabel=ethereum&toLabel=dollar"
url_xrp_usd = "https://www.boursorama.com/bourse/devises/cours/historique/9xXRPUSDSPOT?fromSymbol=XRP&toSymbol=USD&fromLabel=ripple&toLabel=dollar"
url_ltc_usd = "https://www.boursorama.com/bourse/devises/cryptomonnaies/cours/historique/9xLTCUSDSPOT?fromSymbol=LTC&toSymbol=USD&fromLabel=litecoin&toLabel=dollar"
```

```python
def data_array(url):
    request_text = request.urlopen(url).read()
    page = bs4.BeautifulSoup(request_text, "lxml")
    array = page.find('table', {'class': 'c-table'})
    table_body = array.find('tbody')
    rows = table_body.findAll('tr')
    cols = [rows[i].find_all('td') for i in range(len(rows))]
    colonnes = [cols[k] for k in range(len(cols))]
    final_data = np.array([[x[i].text.strip() for i in range(len(x))] for x in colonnes])
    l = ["Date","Fermeture (USD)","Var%","Max","Min","Ouverture(USD)"]
    df = pd.DataFrame(final_data,columns = l)
    return df
```

```python
link = [url_btc_usd,url_eth_usd,url_xrp_usd,url_ltc_usd]
```

```python
for x in link : 
    print(data_array(x))
##Cours jusqu'à la date du 22-Déc
```
